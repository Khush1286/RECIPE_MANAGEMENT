```python
from tkinter import *
import tkinter as tk
from tkinter import PhotoImage
from tkinter import messagebox
import pymysql

top = Tk()
top.title("Login Page")

def hide():
    image1.config(file=r'C:\Users\khush\Pictures\close.png')
    passwordEntry.config(show='*')
    eyeButton.config(command=show)
def show():
    image1.config(file=r'C:\Users\khush\Pictures\openi.png')
    passwordEntry.config(show='')
    eyeButton.config(command=hide)
def signuppage():
    top.destroy()
    %run Signup.ipynb
    
def login_user():
    if  usernameEntry.get() == '' or passwordEntry.get() == '':
        messagebox.showerror('Error', 'All Fields Are Required')
    else:
        try:
            con = pymysql.connect(host='localhost', user='root', password='Khushboo1286', database='login_system')
            mycursor = con.cursor()
            
            query = 'SELECT * FROM table1 WHERE username=%s AND password=%s'
            mycursor.execute(query,(usernameEntry.get(),passwordEntry.get()))
            
            row = mycursor.fetchone()
            if row is None:
                messagebox.showerror('Error', 'Invalid username or password')
            else:
                messagebox.showinfo('Welcome', 'Login is Successful')
            con.close()
            
        except pymysql.Error as e:
            messagebox.showerror('Error', f'Database error: {e}')
    top.destroy()
    %run Mainmenu.ipynb
    
       
canvas_width = 1600
canvas_height = 900
canvas = Canvas(top, width=canvas_width, height=canvas_height)
canvas.pack()

image = PhotoImage(file=r'C:\Users\khush\Pictures\Bgpic1.png')

center_x = canvas_width // 2
center_y = canvas_height // 2

canvas.create_image(center_x, center_y, anchor=CENTER, image=image)
top.image = image

L1=Label(top,text='USER LOGIN',font=('bold',23),bg='burlywood2',fg='black')
L1.place(x=700,y=350)

username=Label(top,text='Username',width=18,font=('bold',14))
username.place(x=700,y=410)

usernameEntry=Entry(top,width=25,font=('bold',11))
usernameEntry.place(x=700,y=440)

password=Label(top,text='Password',width=18,font=('bold',14))
password.place(x=700,y=480)

passwordEntry=Entry(top,width=25,font=('bold',11))
passwordEntry.place(x=700,y=510)



canvas_width1 = 16
canvas_height1 = 9
canvas1 = Canvas(top, width=canvas_width1, height=canvas_height1)
canvas1.pack()

image1 = PhotoImage(file=r'C:\Users\khush\Pictures\openi.png')

center_x = canvas_width1 // 2
center_y = canvas_height1 // 2

canvas1.create_image(center_x, center_y, anchor=CENTER, image=image1)
top.image1 = image1



canvas_width2 = 16
canvas_height2 = 9
canvas2 = Canvas(top, width=canvas_width1, height=canvas_height1)
canvas2.pack()

image2 = PhotoImage(file=r'C:\Users\khush\Pictures\close.png')

center_x = canvas_width2 // 2
center_y = canvas_height2 // 2

canvas2.create_image(center_x, center_y, anchor=CENTER, image=image2)
top.image2 = image2

eyeButton=Button(top,image=image1,width=12,height=12,bg='snow',activebackground='white',cursor='hand2',command=hide)
eyeButton.place(x=886,y=511)  

forgetButton=Button(top,text='Forgot Password?',bd=0,bg='tan1',activebackground='white',cursor='hand2')
forgetButton.place(x=805,y=535)  

loginButton=Button(top,text='LOGIN',font=('bold'),width=13,height=1,command=login_user)
loginButton.place(x=720,y=570)

OrLabel=Label(top,text='-----------------------OR---------------------',font=('bold',12))
OrLabel.place(x=679,y=620)

OrLabel1=Label(top,text='Dont have a account?' ,font=('bold',9),width=20,bg='tan1')
OrLabel1.place(x=679,y=650)

AccButton=Button(top,text='Create new One',font=('bold underline',9),width=15,bg='tan1',activebackground='sandy brown',bd=0,fg='black',cursor='hand2',command=signuppage)
AccButton.place(x=816,y=650)



top.mainloop()
```
